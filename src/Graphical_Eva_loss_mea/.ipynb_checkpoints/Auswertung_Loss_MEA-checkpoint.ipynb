```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('CNN_RESNET_Logger_rgb.csv')
df.head()
```

```python
x = df['epoch']
y1 = df['loss']
y2 = df['val_loss']


plt.plot(x,y1, label='Loss')
plt.plot(x,y2, label='val Loss')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.title('RGB')
plt.legend()
plt.savefig('RESNET_RGB_LOSS.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
x = df['epoch']
y1 = df['mean_absolut_error']
y2 = df['val_mean_absolut_error']


plt.plot(x,y1, label='mean absolut error')
plt.plot(x,y2, label='val mean aboulut error')
plt.xlabel('epochs')
plt.ylabel('MAE')
plt.title('RGB')
plt.legend()
plt.savefig('RESNET_RGB_MAE.png', dpi=300, bbox_inches='tight')
plt.show()
```
